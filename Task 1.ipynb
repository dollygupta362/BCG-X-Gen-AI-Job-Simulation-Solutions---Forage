**Import Important Libraries**

```python
 import pandas as pd
 import matplotlib.pyplot as plt
 import seaborn as sns
 import openpyxl
```

**Now, we will load data.**

```python
df = pd.read_excel(r"C:\Users\dolly\OneDrive\Desktop\10-K Filings.xlsx")
```

**Creating Column 'Revenue Growth (%)'.**

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue ($)'].pct_change() * 100
```

**Fetching dataset after creating the column.**

```python
df
```

**Replacing null values with '0' in column Revenue Growth(%).**

```python
df['Revenue Growth (%)'] = df['Revenue Growth (%)'].fillna(0)
```

**Now fetching the values across the dataset.**

```python
df
```

**Creating Column 'Net Income Growth (%)'.**

```python
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income ($)'].pct_change() * 100
```

**Fetching dataset again.**

```python
df
```

**Replacing null values with '0' in column 'Net Income Growth (%)'.**

```python
df['Net Income Growth (%)'] = df['Net Income Growth (%)'].fillna(0)
```

**Fetching dataset again.**

```python
df
```

**As we have completed the EDA process, we will create some visuals.**

**Plot 1:**
**Total Revenue Trend by Company.**

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x="Fiscal Year", y="Total Revenue ($)", hue="Company", marker="o")
plt.title("Total Revenue Trend")
plt.show()
```

**Insight: Apple consistently leads in revenue, followed by Microsoft, while Tesla lags behind.**

**Plot 2:**
**Net Income Trend by Company**

```python
plt.figure(figsize=(8,5))
sns.lineplot(data=df, x="Fiscal Year", y="Net Income ($)", hue="Company", marker="o")
plt.title("Net Income Trend")
plt.show()
```

**Insight: Apple remains most profitable, Microsoft shows stability, Tesla struggles with declining profits.**

**Plot 3:**
**Assets vs Liabilities (2024)**

```python
plt.figure(figsize=(6,6))
df_2024 = df[df["Fiscal Year"]==2024]
df_2024.set_index("Company")[["Total Assets ($)","Total Liabilities ($)"]].plot(kind="bar")
plt.title("Assets vs Liabilities (2024)")
plt.ylabel("Value ($)")
plt.show()
```

**Insight: Microsoft has the strongest balance sheet, while Apple is highly leveraged.**

**Plot 4:**
**Operating Cash Flow Comparison**

```python
plt.figure(figsize=(6,5))
sns.barplot(data=df, x="Company", y="Cash Flow from Operating Activities ($)", palette="Blues")
plt.title("Operating Cash Flow")
plt.show()
```

**Insight: Apple and Microsoft generate similar strong cash flows, Tesla is far lower.**

**Plot 5:**
**Revenue Growth (%) Comparison**

```python
plt.figure(figsize=(6,5))
sns.barplot(data=df, x="Company", y="Revenue Growth (%)", hue="Fiscal Year")
plt.title("Revenue Growth by Year")
plt.show()
```

**Insight: Tesla showed strong growth in 2022 but flattened later, Apple faced slight decline.**

**Plot 6:**
**Net Income Growth (%) Comparison**

```python
plt.figure(figsize=(6,5))
sns.barplot(data=df, x="Company", y="Net Income Growth (%)", hue="Fiscal Year")
plt.title("Net Income Growth by Year")
plt.show()
```

**Insight: Tesla had a huge profit jump in 2023, while Apple maintained steady small growth.**

**Plot 7:**
**Revenue vs Net Income Scatter**

```python
plt.figure(figsize=(6,5))
sns.scatterplot(data=df, x="Total Revenue ($)", y="Net Income ($)", hue="Company", s=100)
plt.title("Revenue vs Net Income")
plt.show()
```

**Insight: Apple and Microsoft convert revenue into profit more efficiently than Tesla.**
